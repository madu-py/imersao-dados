```python
import numpy as np # Importa a biblioteca numpy, frequentemente usada para operações numéricas, incluindo o tratamento de valores NaN.

#criação de um dataframe de teste para exemplo
df_salarios = pd.DataFrame({ # Cria um novo DataFrame de exemplo chamado 'df_salarios'.
    'nome': ['Ana', 'Bruno', 'Carlos', 'Diana', 'Eduardo'], # Define a coluna 'nome'.
    'salario': [4000, np.nan, 3500, np.nan, 5000] # Define a coluna 'salario' com valores nulos (np.nan).
})

#calcula a media e substitui os nulos pela média e arredonda
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2)) # Cria uma nova coluna 'salario_media' preenchendo os valores nulos da coluna 'salario' com a média dos salários, arredondada para 2 casas decimais.


#calcula a mediana e substitui os nulos pela mediana
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].median()) # Atualiza a coluna 'salario_media', preenchendo os valores nulos da coluna 'salario' com a mediana dos salários.

df_salarios # Exibe o DataFrame 'df_salarios'.
```

```python
df_temperaturas = pd.DataFrame({ # Cria um novo DataFrame de exemplo chamado 'df_temperaturas'.
    'dia': ['Seg', 'Ter', 'Qua', 'Qui', 'Sex'], # Define a coluna 'dia'.
    'temperatura': [30, np.nan, np.nan, 28, 27] # Define a coluna 'temperatura' com valores nulos.
})

#ffill (forward fill) preenche o valor ausente com o valor da linha anterior
df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill() # Cria uma nova coluna 'preenchido_ffill' preenchendo os valores nulos da coluna 'temperatura' usando o método ffill (forward fill), que utiliza o último valor válido observado.

df_temperaturas # Exibe o DataFrame 'df_temperaturas'.
```

```python
df_temperaturas = pd.DataFrame({ # Cria um novo DataFrame de exemplo chamado 'df_temperaturas'.
    'dia': ['Seg', 'Ter', 'Qua', 'Qui', 'Sex'], # Define a coluna 'dia'.
    'temperatura': [30, np.nan, np.nan, 28, 27] # Define a coluna 'temperatura' com valores nulos.
})

#bfill (backward fill) preenche o valor ausente com o valor da linha posterior
df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill() # Cria uma nova coluna 'preenchido_bfill' preenchendo os valores nulos da coluna 'temperatura' usando o método bfill (backward fill), que utiliza o próximo valor válido observado.

df_temperaturas # Exibe o DataFrame 'df_temperaturas'.
```

```python
df_cidades = pd.DataFrame({ # Cria um novo DataFrame de exemplo chamado 'df_cidades'.
    'nome': ['Ana', 'Bruno', 'Carlos', 'Diana', 'Eduardo'], # Define a coluna 'nome'.
    'cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Salvador'] # Define a coluna 'cidade' com valores nulos.
})

#preenchimento com valor fixo
df_cidades['cidade_corrigida'] = df_cidades['cidade'].fillna('Não informado') # Cria uma nova coluna 'cidade_corrigida' preenchendo os valores nulos da coluna 'cidade' com a string 'Não informado'.

df_cidades # Exibe o DataFrame 'df_cidades'.
```

# Desafio Aula 3: Gerando um gráfico interativo dos salários médios de cientista de dados por país

```python
import plotly.express as px # Importa a biblioteca Plotly Express, que é uma interface de alto nível para o Plotly e facilita a criação de gráficos interativos de forma rápida.

# Filtrar para incluir apenas 'Data Scientist'
# Selecionamos as linhas do DataFrame 'df_limpo' onde o valor na coluna 'cargo' é igual a 'Data Scientist'.
# Isso cria um novo DataFrame 'df_ds' contendo apenas os dados para este cargo específico.
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']

# Agrupar por país e calcular a média salarial
# Agrupamos o DataFrame 'df_ds' pela coluna 'residencia' (país).
# Em seguida, calculamos a média da coluna 'usd' (salário em USD) para cada grupo (cada país).
# O .reset_index() converte o resultado do agrupamento de volta para um DataFrame regular, com 'residencia' e 'usd' como colunas.
df_ds_pais = df_ds.groupby('residencia')['usd'].mean().reset_index()

# Ordenar os países pela média salarial (opcional, mas recomendado para visualização)
# Ordenamos o DataFrame 'df_ds_pais' em ordem decrescente ('ascending=False') com base nos valores da coluna 'usd' (média salarial).
# Isso ajuda a visualizar os países com os maiores salários primeiro no gráfico.
df_ds_pais = df_ds_pais.sort_values(by='usd', ascending=False)

# Criar o gráfico de barras interativo com Plotly Express, adicionando personalizações
# Utilizamos a função px.bar() para criar o gráfico de barras:
# - data: Especifica o DataFrame a ser utilizado (df_ds_pais).
# - x: Define a coluna para o eixo x ('residencia' - países).
# - y: Define a coluna para o eixo y ('usd' - média salarial).
# - title: Define o título principal do gráfico.
# - labels: Define rótulos mais amigáveis para os eixos, substituindo os nomes das colunas.
# - color: Define a coluna a ser usada para colorir as barras. Ao usar 'residencia', cada país terá uma cor diferente.
# - template: Define um tema visual para o gráfico. 'plotly_white' é um template com fundo branco e visual limpo.
# - hover_data: Personaliza as informações exibidas ao passar o mouse sobre as barras. ':.2f' formata o salário com duas casas decimais.
fig = px.bar(df_ds_pais,
             x='residencia',
             y='usd',
             title='Média Salarial de Data Scientists por País (USD)',
             labels={'residencia': 'País', 'usd': 'Média Salarial Anual (USD)'},
             color='residencia',
             template='plotly_white',
             hover_data={'residencia': True, 'usd': ':.2f'}
            )

# Atualizar layout para um visual mais moderno e profissional
# Utilizamos fig.update_layout() para fazer ajustes finos no layout do gráfico:
# - xaxis_title: Define o título do eixo x.
# - yaxis_title: Define o título do eixo y.
# - showlegend=False: Remove a legenda, já que as cores representam os países que já estão no eixo x.
# - title_font_size, xaxis_title_font_size, yaxis_title_font_size: Ajustam o tamanho da fonte para os títulos.
# - hovermode='closest': Define o comportamento do tooltip ao passar o mouse.
fig.update_layout(
    xaxis_title='País',
    yaxis_title='Média Salarial Anual (USD)',
    showlegend=False,
    title_font_size=20,
    xaxis_title_font_size=14,
    yaxis_title_font_size=14,
    hovermode='closest'
)

# Exibir o gráfico
# Este comando renderiza o gráfico interativo.
fig.show()
```
